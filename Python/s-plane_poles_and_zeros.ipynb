This code plots the poles and zeros of a transfer function, in the s-Domain

The 's' elements in the transfer function are complex numbers.
Poles are located where the denominator of the transfer function is zero (infinite response)
Zeros are located where the numerator of the transfer function is zero.
As a simple example, a one-pole filter has the following transfer function: H(s) = 1 / (1 + s/p)
This has one real root in the denominator, at s = âˆ’p.

Numerator and denominator polynomials can be specified, below and the Pole-Zero map plotted

As additional information, here is a great explanation as to why the complex poles and zeros must always be complex conjugate pairs: https://dsp.stackexchange.com/questions/68930/is-it-a-mandatory-condition-that-complex-poles-and-zeros-should-always-exist-as



```python
# If you would like to run this in Colab then it doesn't include the control package by
# default so uncomment this line of code, to install it:

# ! pip install control
```

```python
import numpy as np
import control
num = np.array([3, 2, 1])
den = np.array([1, 2, 3])

H = control.tf(num , den)
print ('H(s) =', H)
z = control.zero(H)
print ('z =', z)
p = control.pole(H)
print ('p =', p)
control.pzmap(H)
```
